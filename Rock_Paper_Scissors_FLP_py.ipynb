```python
import random

def play_rock_paper_scissors():
    # Data: Define the valid choices
    choices = ["rock", "paper", "scissors"]

    print("Welcome to Rock-Paper-Scissors!")
    print("Type 'quit' to stop playing.\n")

    while True:
        # Built-in Function: input() to get user input
        user_input = input("Enter your choice (rock, paper, scissors): ").lower()

        # Check if user wants to quit
        if user_input == 'quit':
            print("Thanks for playing! Goodbye.")
            break

        # Validate user input
        if user_input not in choices:
            print("Invalid choice! Please enter rock, paper, or scissors.\n")
            continue

        # Data: Generate a random choice for the computer
        computer_choice = random.choice(choices)

        print(f"Computer chose: {computer_choice}")

        # Model: Use conditional statements to determine the winner
        if user_input == computer_choice:
            print(f"Both players selected {user_input}. It's a tie!")

        elif user_input == "rock":
            if computer_choice == "scissors":
                print("Rock smashes scissors! You win!")
            else:
                print("Paper covers rock! You lose.")

        elif user_input == "paper":
            if computer_choice == "rock":
                print("Paper covers rock! You win!")
            else:
                print("Scissors cuts paper! You lose.")

        elif user_input == "scissors":
            if computer_choice == "paper":
                print("Scissors cuts paper! You win!")
            else:
                print("Rock smashes scissors! You lose.")

        print("-" * 30) # Separator for readability

if __name__ == "__main__":
    play_rock_paper_scissors()
```
